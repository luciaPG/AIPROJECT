So, as result, there is no need to standalize the name(string), but it's better to standalize the number, as 1 acre = 43560 sqft

```python
df_train.loc[df_train['lot_size_units'] == 'acre', 'lot_size'] *= 43560 #Change the acre to sqft
df_train.loc[df_train['size_units'] == 'acre', 'size'] *= 43560 #Change the acre to sqft
df_train = df_train.drop('lot_size_units', axis=1) #Delet the units line because we dont need it when its all in sqft
df_train = df_train.drop('size_units', axis=1) #Delet the units line because we dont need it when its all in sqft
df_train
```

```python
df_test.loc[df_test['lot_size_units'] == 'acre', 'lot_size'] *= 43560 #Change the acre to sqft
df_test.loc[df_test['size_units'] == 'acre', 'size'] *= 43560 #Change the acre to sqft
df_test = df_test.drop('lot_size_units', axis=1) #Delet the units line because we dont need it when its all in sqft
df_test = df_test.drop('size_units', axis=1) #Delet the units line because we dont need it when its all in sqft
df_test
```

Then, we are going to delete the same lines(if there is)

```python
df_train = df_train.drop_duplicates(keep='last')
df_train
```

```python
df_test = df_test.drop_duplicates(keep='last')
df_test
```

Both datasets there is a problem of NaN variable

```python
df_train = df_train.dropna().reset_index(drop=True) #Drop Nan andd reset the index
df_test = df_test.dropna().reset_index(drop=True)
print(df_train)
print(df_test)
```
